```python
from pathlib import Path
import pandas as pd

from constants import TOXICITY_CLASSIFICATIONS_H5, TEXTS_DIR
```

```python
store = pd.HDFStore(TOXICITY_CLASSIFICATIONS_H5, mode='r')
df = store['df']
```

```python
toxic = df[df['toxic'] > 0.5].sort_values(by='toxic', ascending=False)
```

```python
toxic.head(10)
```

```python
toxic['text'] = toxic['filename'].apply(lambda fn: (TEXTS_DIR / fn).read_text())
toxic.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(toxic['text'], test_size=0.2, random_state=42)
```

```python
print('\n'.join(train), file=open('gpt2.train.txt', 'w'))
print('\n'.join(test), file=open('gpt2.val.txt', 'w'))
```
